```python
soup = BeautifulSoup(res.text, 'html.parser')
soup
```

```python
print(soup.prettify())
```

```python
# 物件名を取得
houses = soup.find_all('div', class_='cassetteitem')
house = houses[0]
house_name = house.find('div', class_='cassetteitem_content-title').text
house_name
```

```python
# 徒歩何分かを取得
house_min = house.find('li', class_='cassetteitem_detail-col2').text
house_min = int(house_min.split('\n')[1].split('歩')[1].split('分')[0])
house_min
```

```python
# 家賃を取得
house_price_box = house.find('table', class_='cassetteitem_other')
house_price_room = house_price_box.find_all('tr', class_='js-cassette_link')[0]
house_price = house_price_room.find('span', class_='cassetteitem_other-emphasis ui-text--bold').text
house_price = float(house_price.replace('万円', ''))
house_price
```

```python
# 辞書型にまとめる
datum = {}
datum['name'] = house_name
datum['min'] = house_min
datum['price'] = house_price

datum
```

```python
data = []
for number in range(1, 40):
    url = f'https://suumo.jp/jj/chintai/ichiran/FR301FC001/?ar=030&bs=040&ra=013&rn=0090&cb=0.0&ct=9999999&mb=0&et=9999999&cn=9999999&shkr1=03&shkr2=03&shkr3=03&shkr4=03&sngz=&po1=25&pc=50&page={number}'
    res = requests.get(url)

    soup = BeautifulSoup(res.text, 'html.parser')

    

    # 物件名を取得
    houses = soup.find_all('div', class_='cassetteitem')
    for house in houses:
        house_name = house.find('div', class_='cassetteitem_content-title').text

        # 徒歩何分かを取得
        house_min = house.find('li', class_='cassetteitem_detail-col2').text
        house_min = int(house_min.split('\n')[1].split('歩')[1].split('分')[0])

        # 家賃を取得
        house_price_box = house.find('table', class_='cassetteitem_other')
        house_price_room = house_price_box.find_all('tr', class_='js-cassette_link')[0]
        house_price = house_price_room.find('span', class_='cassetteitem_other-emphasis ui-text--bold').text
        house_price = float(house_price.replace('万円', ''))

        # 辞書型にまとめる
        datum = {}
        datum['name'] = house_name
        datum['min'] = house_min
        datum['price'] = house_price
        data.append(datum)
    sleep(1)
```

```python
data
```

```python
df = pd.DataFrame(data)
df
```

```python
df.to_csv('suumo1.csv')
```
